## Manipulation de dictionnaires (GeoJson)

```python
# réimport au cas où (normalement c'est déjà chargé plus haut)
import json
# rechargement du fichier (là encore au cas où)
data = json.load(open('./data/arbresremarquablesparis.geojson', 'r',encoding='utf8'))
```

Créer une liste avec les valeurs uniques de l'attribut 'libellefrancais'

```python
# votre code ici
liste_valeurs = [feature["properties"]["libellefrancais"] for feature in data["features"]]
set_valeurs = set(liste_valeurs)
liste_finale = list(set_valeurs)
liste_finale.sort()
# affichage pour le notebook (sinon utiliser print())
liste_finale
```

Créer une liste de features correspondant aux seuls Chênes (libellefrancais)

```python
# votre code ici
[feature for feature in data["features"] if feature["properties"]["libellefrancais"] == 'Chêne']
```

### Ajout d'attribut dans un dictionnaire

Inspecter l'attribut arrondissement en listant les valeurs uniques

```python
# Votre code ici 

liste_valeurs = [feature["properties"]["arrondissement"] for feature in data["features"]]
set_valeurs = set(liste_valeurs)
liste_finale = list(set_valeurs)
liste_finale.sort()
liste_finale
```


Puis modifier les objets dans le GeoJson pour créer un champ num_arrondissement qui ne contiendra que le numéro de l'arrondissement au format int

```python
# Votre code ici 
for feature in data["features"] :
    temp = feature["properties"]["arrondissement"].split(" ")[1]
    temp = temp.replace("E", "")
    if temp.isnumeric() :
        arr = int(temp)
    else :
        arr = 0
    feature["properties"]["arrondissement_num"] = arr

data
```
